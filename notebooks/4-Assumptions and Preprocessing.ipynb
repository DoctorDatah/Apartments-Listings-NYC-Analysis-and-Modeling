# Probability Distribution funtion

**PDF Rent**

```python
sns.distplot(listings['rent'] )

# Above $10000 are outliers 
```

```python
# Getting 96% Quantile
q = listings['rent'].quantile(0.96)
q
```

```python
listings_b = listings[listings['rent']<q]
listings_b.info()
```

```python
sns.distplot(listings_b['rent'] )
```
